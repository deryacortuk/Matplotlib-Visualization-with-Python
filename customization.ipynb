```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,7)
y=np.arange(2,14,2)
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
axes =fig.add_axes([0,0,1,1])
axes.plot(x,y**2,color ="blue",lw=3,linestyle="--",marker="o",markersize=8,markerfacecolor="green",
markeredgecolor="orange",markeredgewidth=5)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
axes =fig.add_axes([0,0,1,1])
axes.plot(x*2,y**2,color ="green",lw=3,linestyle="--",marker="o",markersize=8,markerfacecolor="blue",
markeredgecolor="purple",markeredgewidth=5)

axes.set_xlim(0,30)

axes.set_ylim(0,170)
```
