```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,8)
y=np.arange(1,14,2)
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

for i in axes:
    i.plot(x,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

axes[0].plot(x**2,y**3)
axes[1].plot(y**2,x*3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
axes[0].plot(x*x,y**3,color="purple")
axes[0].set_title("first")
axes[1].plot(x,y*y,color="blue")
axes[1].set_title("second")

plt.tight_layout()

```

```python
fig=plt.figure(figsize=(17,7))
axes =fig.add_axes([0,0,1,1])
axes.plot(x*y,y**3,color="orange")
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,5))
axes[0].plot(x**2+y**2,y**x,color="red")
axes[0].set_title("first")
axes[1].plot(x**0.2,y**2,color="purple")
axes[1].set_title("second")
plt.tight_layout()
```

```python
fig =plt.figure(figsize=(9,9))
axes =fig.add_axes([0,0,1,1])
axes.plot(x*x,x**3,color="blue",label="blue")
axes.plot(x*x*7,x**3,color="#396cec",label="siben")
axes.plot(x,x**3,color="red",label="red")
axes.plot(x**3,x*x,color="black",label="hue")
axes.plot(x**3,x**3,color="#e22e5d",label="pink")
axes.legend()

```
