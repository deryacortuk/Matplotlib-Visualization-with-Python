```python
import numpy as np
import matplotlib.pyplot as plt

phi = np.linspace(0, 2*np.pi, 100)
r = 1
x = r*np.cos(phi)
y = r*np.sin(phi)
plt.plot(x,y,color ="blue",lw=4,linestyle="--",marker="o",markersize=8,markerfacecolor="purple",
markeredgecolor="yellow",markeredgewidth=5)
plt.axis("equal")
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y =np.sin(x)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(11,7))


ax[0].plot(x, y,color ="blue",lw=3)
ax[0].set(xlabel='radians', ylabel='Sin(x)',title='Sine Plot')
ax[0].grid()

z = np.cos(x)

ax[1].plot(x, z,color="red",lw=3)
ax[1].set(xlabel='radians', ylabel='(cosx)',title='Cosine Plot')
ax[1].grid()

plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
t =np.tan(x)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(13,9))


ax[0].plot(x, t,color ="green",lw=3)
ax[0].set(xlabel='radians', ylabel='Tan(x)',title='Tane Plot')
ax[0].grid()


k=1/(np.tan(x))
ax[1].plot(x, k,color="purple",lw=3)
ax[1].set(xlabel='radians', ylabel='Cot(x)',title='Cote Plot')
ax[1].grid()

plt.show()
```
