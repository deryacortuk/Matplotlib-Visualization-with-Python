```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(1,23,3)
y=np.arange(3,19,2)
z=x*y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x*y,y**2,color="blue",lw=3,linestyle="-.",marker="o",markerfacecolor="red")
axes.set_xlabel("x values")
axes.set_ylabel("y values")
axes.set_title("x and y graph")

plt.show()
```

```python
fig = plt.figure(figsize=(9,8))

axes1=fig.add_axes([0,0.1,1,1])
axes2=fig.add_axes([0.1,0.5,0.4,0.5])

axes1.plot(y**2,z*x,color="green",lw=3,linestyle="--",marker="o",markersize=7,markerfacecolor="purple",
markeredgecolor="blue",markeredgewidth=5)
axes1.set_title("Outer Graph")

axes2.plot(x*z,x**2, color="orange",lw=3,marker="o",markersize=5,markerfacecolor="blue",
markeredgecolor="purple",markeredgewidth=5)
axes.set_title("Inner Graph")
```

```python
fig,axes =plt.subplots(nrows =1,ncols=2,figsize=(9,7))

axes[0].plot(x,y,color="red",lw=3,ls="--",marker="o",markerfacecolor="blue",markersize=7)
axes[0].set_xlabel("x values")
axes[0].set_ylabel("y values")
axes[0].set_title("x and y graph")

axes[1].plot(x**2,z**3,color="green",lw=3,ls="-.",marker="o",markerfacecolor="purple",markersize=9)
axes[1].set_xlabel("x values")
axes[1].set_ylabel("y values")
axes[1].set_title("x and z graph")
```
